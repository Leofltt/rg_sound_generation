```python
%matplotlib inline

import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

from config import get_config

plt.style.use("dark_background")
```

```python
conf = get_config()

features = conf.get("all_features")
```

```python
f = features[0]

df1 = pd.read_csv(f"logs/all/{f}.csv")
df2 = pd.read_csv(f"logs/{f}.csv")

df1.head()
```

```python
plt.figure(figsize=(16, 16))

for i in range(0, 9):
    f= features[i]
    df1 = pd.read_csv(f"logs/all/{f}.csv")
    df2 = pd.read_csv(f"logs/{f}.csv")

    plt.subplot(3, 3, i + 1)
    plt.plot(df1.iloc[:, 4], label="val acc")
    plt.plot(df1.iloc[:, 1], label="train acc")
    plt.plot(df2.iloc[:, 4], label="val acc (lucas)")
    plt.plot(df2.iloc[:, 1], label="train acc (lucas)")
    plt.legend()
    plt.title(f)
    plt.ylim([0., 1.])
plt.savefig("acc.png")
```

```python
plt.figure(figsize=(16, 16))

for i in range(0, 9):
    f= features[i]
    df1 = pd.read_csv(f"logs/all/{f}.csv")
    df2 = pd.read_csv(f"logs/{f}.csv")

    plt.subplot(3, 3, i + 1)
    plt.plot(df1.iloc[:, 5], label="val loss")
    plt.plot(df1.iloc[:, 2], label="train loss")
    plt.plot(df2.iloc[:, 5], label="val loss (lucas)")
    plt.plot(df2.iloc[:, 2], label="train loss (lucas)")
    plt.legend()
    plt.title(f)
plt.savefig("loss.png")
```
